```python
import numpy as np
import time
import matplotlib.pyplot as plt
data=np.loadtxt("DATA.txt")
data1=np.loadtxt("DATA1.txt")
def draw_figure(y,y1,i):
    Z=np.ones((len(y),1))*100/12
    x = np.linspace(0, len(y)/100, len(y))
    x1 = np.linspace(0, len(y1)/100, len(y1))
    plt.figure()    # 定义一个图像窗口
    plt.plot(x,Z,color='r')
    plt.plot(x, y/12,linewidth=1)#,color='r') # 绘制曲线 y1
    plt.plot(x1, y1/12,linewidth=1)#,color='r') # 绘制曲线 y1
    # plt.xlim(0,20)
    # plt.ylim(0,200)
    plt.legend(["Threshohld","Rigid body","Orgin tissue"])
    plt.xlabel("Times(s)",fontsize=12,   )
    plt.ylabel("Current of motor (mA)",fontsize=12,   )
    # 设置横轴精准刻度
    # plt.xticks([0,5,10,15,20],fontsize=12,)
    # # 设置纵轴精准刻度
    # plt.yticks(np.linspace(0,120,2),fontsize=12,)
    # plt.axis('off')
    plt.savefig(str(i)+".png")
    plt.show()

    # cv2.namedWindow('image', cv2.WINDOW_NORMAL) 
    # img = cv2.imread("datafig/fig"+str(i)+".png")
    # cv2.imshow('image',img) 
    # cv2.waitKey(0)

draw_figure(data,data1,100)
```
